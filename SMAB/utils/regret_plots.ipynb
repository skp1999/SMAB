```python
data = [eval(line) for line in open('experiments_sst/thompson_binary_sst_english_llama2_13B1_regret_200k/regret_per_iteration.json','r')]
```

```python
values = []
print(len(data[0]))
for key, value in data[0].items():
    values.append(value)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
time_steps = np.arange(1, 20001)  # Time steps from 1 to 100
values = values[:20000]  # Cumulative sum of random values for the example

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(time_steps, values, label='Cumulative Value', color='b')

# Adding labels and title
plt.xlabel('Time Steps')
plt.ylabel('Values')
plt.title('Line Chart of Values Over Time')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
np.random.seed(42)
# Example data
time_steps = np.arange(1, 20001)  # Time steps from 1 to 100
values = values[:20000]  # Cumulative sum of random values for the example

# Moving average function to smooth data
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

# Apply moving average with a window size of 5
smoothed_values = moving_average(values, window_size=4000)

# Plot original and smoothed data
plt.figure(figsize=(8, 6))
# plt.plot(time_steps, values, label='Original Data', alpha=0.5, color='gray')
plt.plot(time_steps[:len(smoothed_values)], smoothed_values, label='Smoothed Data', color='b')

# Adding labels and title
plt.xlabel('Time Steps')
plt.ylabel('Values')
plt.title('Smooth Line Chart using Moving Average')
plt.legend()

# Show the plot
# plt.grid(True)
plt.show()
```
